```python
feature_columns = X_test_encoded.columns.drop('test')

# Store columns with their percentage of correct predictions
predictor_stats = []

# Iterate through each one-hot encoded column
for col in feature_columns:
    # Select rows where the feature is 1
    subset = X_test_encoded[X_test_encoded[col] == 1]

    # If subset is empty, skip
    if len(subset) == 0:
        continue

    # Count the most common target value
    value_counts = subset['test'].value_counts()
    most_common_val = value_counts.idxmax()
    most_common_count = value_counts.max()
    total = len(subset)

    # Calculate percentage of records correctly predicting the most common target value
    percentage = (most_common_count / total) * 100

    # Append stats
    predictor_stats.append((col, most_common_val, most_common_count, total, percentage))

# Sort by percentage descending
predictor_stats.sort(key=lambda x: x[4], reverse=True)

# Print
print("Column | Most Common Target | Count | Total | Percentage")
for col, val, count, total, pct in predictor_stats:
    print(f"{col} | {val} | {count} | {total} | {pct:.2f}%")
```

**Key Findings**

**As observed during EDA, certain features such as odor, gill size, stalk surface texture, bruises, and spore print color showed strong associations with mushroom edibility. Specifically, mushrooms with no odor, narrow gills, and smooth stalk surfaces tended to be edible, while those with foul or pungent odors, crowded gills, or chocolate spore prints were often poisonous.**

**The trained model reinforced these observations. According to the feature importance chart, odor_none emerged as the most predictive feature, strongly indicating edibility, while odor_foul is among the top contributors signaling poisonousness.**

**While habitat and population have a comparatively lower impact, they still contribute to classification and align with the patterns detected during EDA.**

**The model, emphasizes that odor, gill characteristics, stalk surface texture, and spore print color are the most decisive features in predicting whether a mushroom is edible or poisonous.**
